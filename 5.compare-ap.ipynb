```python
import pandas as pd
import plotly.express as px
```

```python
# Read the results

matric = pd.read_csv('output/matric-ap.csv')
non_vectorized = pd.read_csv('output/non-vectorized-ap.csv')
vectorized = pd.read_csv('output/vectorized-ap.csv')
```

```python
# Merge data frames

matric_non_vectorized = matric.merge(non_vectorized, on =['Metadata_Perturbation', 'Cell_Time', 'Metadata_broad_sample'], how='inner')
matric_vectorized = matric.merge(vectorized, on =['Metadata_Perturbation', 'Cell_Time', 'Metadata_broad_sample'], how='inner')
```

## Compare APs

### Matric vs. Non-vectorized

```python
fig = px.scatter(
    matric_non_vectorized,
    x="mAP_x",
    y="mAP_y",
    facet_col="Cell_Time",
    facet_row="Metadata_Perturbation",
    trendline="ols",
)

fig.update_traces(marker=dict(size=3))
fig.update_layout(
    xaxis1=dict(title="mAP<br>matric"),
    xaxis2=dict(title="mAP<br>matric"),
    xaxis3=dict(title="mAP<br>matric"),
    xaxis4=dict(title="mAP<br>matric"),
    yaxis1=dict(title="mAP matric<br>non-vectorized"),
    yaxis5=dict(title="mAP matric<br>non-vectorized"),
    yaxis9=dict(title="mAP matric<br>non-vectorized"),
)

fig.for_each_annotation(
    lambda a: a.update(text=a.text.replace("Metadata_Perturbation=", ""))
)
fig.for_each_annotation(lambda a: a.update(text=a.text.replace("Cell_Time=", "")))
fig.show("png")
```

```python
# R-squared values

for i in range(0,12):
    print(px.get_trendline_results(fig).px_fit_results.iloc[i].rsquared)
```
